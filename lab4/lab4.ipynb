<H3> Golden Ratio </H3>

```python
import numpy as np
import math
eps = 10**(-2)
def method_goldenRatio(f, b: float = 1, a: float = 0, e = eps * 1e-1):
    tau = (math.sqrt(5) + 1) / 2
    n_f = 0
    Ak, Bk = a, b
    lk = Bk - Ak
    Xk1 = Bk - (Bk - Ak) / tau
    Xk2 = Ak + (Bk - Ak) / tau
    
    y1, y2 = f(Xk1), f(Xk2)
    n_f += 2

    while lk > e:
        if y1 < y2:
            Bk = Xk2
            Xk2 = Xk1
            Xk1 = Ak + Bk - Xk2
            y2 = y1
            y1 = f(Xk1)
            n_f += 1
        else:
            
            Ak = Xk1
            Xk1 = Xk2
            Xk2 = Ak + Bk - Xk1
            y1 = y2
            y2 = f(Xk2)
            n_f += 1
        lk = Bk - Ak
    return (Ak + Bk) / 2, n_f
```

<H3> Newton

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from sympy import Symbol, lambdify, diff, derive_by_array
from scipy import optimize

x0 = [-1, -2]
e = 10 ** (-2)

def newton_method(foo, eps = e, x = np.array(x0), nu: float = 0.8, omega: float = 0.5,  flag: int = 0, flag1: int = 0):

    n_f = 0
    n_g = 0
    n_hess = 0
    n_d = 0

    kappa_k = 1

    x = np.array(x)

    x_, y_ = Symbol('x'), Symbol('y')
    grad_f = lambdify((x_, y_), derive_by_array(foo(x_, y_), (x_, y_)), 'numpy')
    
    
    f = lambda x: foo(x[0], x[1])
    
    grad = lambda x: grad_f(x[0], x[1])

    anti_grad = lambda x: -grad_f(x[0], x[1])
    # считаем Гессиан
    h = lambdify((x_, y_), [[diff(foo(x_, y_), x_, 2), diff(foo(x_, y_), x_, y_)], [diff(foo(x_, y_), y_, x_), diff(foo(x_, y_), y_, 2)]], 'numpy')
    H = lambda x: np.array(h(x[0],x[1]))    

    xk = [[],[]]    
    w_norm = []   
    
    while True:      
        
        xk[0].append(x[0])
        xk[1].append(x[1])
        
        
        # в точку x k-1 находим градиент и матрицу Гессе
        wk = anti_grad(x)
        n_g += 1        
        w_norm.append(np.linalg.norm(wk))
        

        # если модуль градиетнта x k-1 меньше eps выходим из цикла
        if w_norm[-1] < eps:            
            break
        hess = H(x)
        n_hess += 1
        # количество итераций
        n_d += 1

        
        if not np.all(np.linalg.eigvals(hess) > 0):
            eta = 0
            while not np.all(np.linalg.eigvals(hess) > 0):
                eta += 0.1
                hess = hess + eta * np.eye(hess.ndim)
        
        pk = np.linalg.solve(hess, wk)      
            
        if flag == 0: # базовый метод Ньютона            
            x = x + pk

        elif flag == 1: # модификация метода Ньютона с наискорейшим спуском            
            
            phi_k = lambda kappa: f(x + kappa * pk)
            temp = 0

            # if flag1 == 0:
            #     kappa_k, temp = method_goldenRatio(phi_k, 0.5, 0, eps/1000)
            # elif flag1 == 1:
            #     kappa_k, n, temp = optimize.golden(phi_k, brack = (0, 0.5), full_output=True)      
        
            kappa_k, temp = method_goldenRatio(phi_k)
            x = x + kappa_k * pk
            n_f += temp
            
        elif flag == 2: # модификация метода Ньютона с дроблением шага (Макгвард)
            
            x_cur = x + kappa_k * pk
            fk = f(x)
            n_f += 1
            kappa_k = 0.8
            while fk - f(x_cur) <= omega * kappa_k * np.dot(-wk, pk):
                kappa_k *= nu
                x_cur = x + kappa_k * pk
                n_f += 1
            x = x_cur
        
    return x, xk, w_norm, n_f, n_g, n_d, n_hess
```
